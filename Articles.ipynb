```python
# Import Libraries
import pandas as pd
import numpy as np
import re
```

```python
data_frame = pd.read_csv("articles.csv",encoding='latin1')
```

```python
data_frame.head()
```

```python
# Removing HTML tags from Article.Description
def remove_tags(string):
    """
    This function is used to remove html tags in the string.
    Parameter : string
    return : string
    """
    result = re.sub('<.*?>','',string)
    return result
```

```python
data_frame['Article.Description'] = data_frame['Article.Description'].apply(lambda cw : remove_tags(cw))
```

```python
data_frame['Full_Article'] = data_frame['Full_Article'].apply(lambda cw : remove_tags(cw))
```

```python
# Head of data frame after Removing HTML tags
data_frame.head()
```

```python
# creating new column Preprocessed_Text
data_frame['Preprocessed_Text'] = data_frame['Heading'].astype(str) +" "+ data_frame['Article.Description'].astype(str) +" "+ data_frame['Full_Article'] 
```

```python
# head of data frame after adding Preprocessed_Text column
data_frame.head()
```

```python
# Import stopwords with nltk.
from nltk.corpus import stopwords
stop = stopwords.words('english')
```

```python
data_frame['Preprocessed_Text'] = data_frame['Preprocessed_Text'].apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
```

```python
# head of data frame after removing Stop words
data_frame.head()
```

```python
# Function To remove Punctuations
import string
def remove_punctuations(text):
    for punctuation in string.punctuation:
        text = text.replace(punctuation, '')
    return text
```

```python
data_frame['Preprocessed_Text'] = data_frame['Preprocessed_Text'].apply(remove_punctuations)
```

```python
# head of data frame after removing Punctuations
data_frame.head()
```

```python
# Removing Trailing And Leading Spaces
data_frame['Preprocessed_Text'] = data_frame['Preprocessed_Text'].apply(lambda x: x.strip())
```

```python
data_frame.head()
```

```python
# Feature Engineering and Classifier Model
```

```python
from sklearn import model_selection, preprocessing, linear_model, naive_bayes, metrics, svm
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn import decomposition, ensemble
from sklearn.metrics import confusion_matrix 
from sklearn.metrics import accuracy_score 
from sklearn.metrics import classification_report
```

```python
text = data_frame['Preprocessed_Text']
y = data_frame['Article_Type']
```

```python
# split the dataset into training and validation datasets 
train_x, valid_x, train_y, valid_y = model_selection.train_test_split(text, y)
```

```python
# label encode the target variable 
encoder = preprocessing.LabelEncoder()
encoder.fit(train_y)
```

```python
encoder_name_mapping = dict(zip(encoder.classes_, encoder.transform(encoder.classes_)))
```

```python
print(encoder_name_mapping)
```

```python
train_y = encoder.fit_transform(train_y)
valid_y = encoder.fit_transform(valid_y)
```

```python
# create a count vectorizer object 
count_vect = CountVectorizer(analyzer='word', token_pattern=r'\w{1,}')
count_vect.fit(text)
```

```python
# transform the training and validation data using count vectorizer object
xtrain_count =  count_vect.transform(train_x)
xvalid_count =  count_vect.transform(valid_x)
```

```python
# word level tf-idf
tfidf_vect = TfidfVectorizer(analyzer='word', token_pattern=r'\w{1,}', max_features=5000)
tfidf_vect.fit(text)
xtrain_tfidf =  tfidf_vect.transform(train_x)
xvalid_tfidf =  tfidf_vect.transform(valid_x)
```

```python
def train_model(classifier, feature_vector_train, label, feature_vector_valid):
    # fit the training dataset on the classifier
    classifier.fit(feature_vector_train, label)
    
    # predict the labels on validation dataset
    predictions = classifier.predict(feature_vector_valid)
    
    return metrics.accuracy_score(predictions, valid_y)
```

```python
# Naive Bayes on Word Level TF IDF Vectors
accuracy = train_model(naive_bayes.MultinomialNB(), xtrain_tfidf, train_y, xvalid_tfidf)
print ("NB, WordLevel TF-IDF: ", accuracy)
classifier = naive_bayes.MultinomialNB()

model = classifier.fit(xtrain_tfidf,train_y)

pred_y = model.predict(xvalid_tfidf)
results = confusion_matrix(valid_y, pred_y) 
print ('Confusion Matrix :')
print(results) 
print ('Accuracy Score :',accuracy_score(valid_y, pred_y)) 
```

```python
# Linear Classifier on Word Level TF IDF Vectors
accuracy = train_model(linear_model.LogisticRegression(), xtrain_tfidf, train_y, xvalid_tfidf)
print ("LR, WordLevel TF-IDF: ", accuracy)
classifier = linear_model.LogisticRegression()

model = classifier.fit(xtrain_tfidf,train_y)

pred_y = model.predict(xvalid_tfidf)
results = confusion_matrix(valid_y, pred_y) 
print ('Confusion Matrix :')
print(results) 
print ('Accuracy Score :',accuracy_score(valid_y, pred_y))
```

```python
# SVM on Word Level TF IDF Vectors
accuracy = train_model(svm.SVC(), xtrain_tfidf, train_y, xvalid_tfidf)
print ("SVM, WordLevel TF-IDF: ", accuracy)
classifier = svm.SVC()

model = classifier.fit(xtrain_tfidf,train_y)

pred_y = model.predict(xvalid_tfidf)
results = confusion_matrix(valid_y, pred_y) 
print ('Confusion Matrix :')
print(results) 
print ('Accuracy Score :',accuracy_score(valid_y, pred_y))
```

```python
# RF on Word Level TF IDF Vectors
accuracy = train_model(ensemble.RandomForestClassifier(), xtrain_tfidf, train_y, xvalid_tfidf)
print ("RF, WordLevel TF-IDF: ", accuracy)
classifier = ensemble.RandomForestClassifier()

model = classifier.fit(xtrain_tfidf,train_y)

pred_y = model.predict(xvalid_tfidf)
results = confusion_matrix(valid_y, pred_y) 
print ('Confusion Matrix :')
print(results) 
print ('Accuracy Score :',accuracy_score(valid_y, pred_y))
```
